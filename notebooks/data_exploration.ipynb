```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python

dev_df = pd.read_csv('../data/assignment1_dev_set.csv')
val_df = pd.read_csv('../data/assignment1_val_set.csv')

# Display basic info of the dataframe, their statistics and missing values

display(dev_df.info())
display(dev_df.head())
display(dev_df.describe())
print("dev set missing values: " + str(dev_df.isnull().sum().sum()))

display(val_df.info())
display(val_df.head())
display(val_df.describe())
print("val set missing values: " + str(val_df.isnull().sum().sum()))
```

```python
# Check for columns containing only one value

for column in dev_df.columns:
    num_unique = dev_df[column].nunique()
    if num_unique == 1:
        print(f"dev: Column '{column}' has only 1 unique value: {dev_df[column].iloc[0]}")

for column in val_df.columns:
    num_unique = val_df[column].nunique()
    if num_unique == 1:
        print(f"val: Column '{column}' has only 1 unique value: {val_df[column].iloc[0]}")
```

```python
# Drop the irrelevant columns
dev_df_2 = dev_df.drop(columns=['Unnamed: 0', 'Project ID', 'Experiment type', 'Disease MESH ID'])


val_df_2 = val_df.drop(columns=['Unnamed: 0', 'Project ID', 'Experiment type', 'Disease MESH ID'])

display(dev_df_2.head())
display(val_df_2.head())
```

```python
# Use LabelEncoder to convert categorical columns to numerical values

# label_encoder = LabelEncoder()
# label_encoder.classes_ = np.array(['Male', 'Female'])
# dev_df_3 = dev_df_2
# for column in dev_df_2.select_dtypes(include=['object']).columns:
#     dev_df_3[column] = label_encoder.fit_transform(dev_df_2[column])



# val_df_3=val_df_2
# for column in val_df_2.select_dtypes(include=['object']).columns:
#     val_df_3[column] = label_encoder.fit_transform(val_df_2[column])

dev_df_3 = dev_df_2
dev_df_3['Sex'] = dev_df_3['Sex'].map({'Male': 0, 'Female': 1})

val_df_3=val_df_2
val_df_3['Sex'] = val_df_3['Sex'].map({'Male': 0, 'Female': 1})

# display(dev_df_3.info())
# display(dev_df_3.head())

# display(val_df_3.info())
# display(val_df_3.head())
```

```python
scaler = StandardScaler()
col_scale = [x for x in dev_df_3.columns if x not in ['BMI', 'Sex']]

dev_df_3[col_scale] = scaler.fit_transform(dev_df_3[col_scale])

val_df_3[col_scale] = scaler.fit_transform(val_df_3[col_scale])


display(dev_df_3.info())
display(dev_df_3.head())

display(val_df_3.info())
display(val_df_3.head())

dev_df_3.to_csv('../data/development_final_data.csv.csv', index=False)  
val_df_3.to_csv('../data/evaluation_final_data.csv', index=False)  
```
